# **01: Ordenar o gr√°fico countplot e melhorar a visualiza√ß√£o alterando o tamanho da fonte.**

Preferi por ordenar o gr√°fico em ordem crescente, j√° que por padr√£o ele ordena da forma decrescente.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns = mapa, inplace = True)
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
sns.set()
plt.figure(figsize = (8, 6))

ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show() ###Retira algumas linhas espec√≠ficas que ficam em cima do gr√°fico
```

```python
sns.set()
plt.figure(figsize = (8, 6))

ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = reversed(dados['composto'].value_counts().index[0:5]))
###Usar o "order" sem passar o "reversed" antes ir√° mostrar um gr√°fico em order decrescente 
ax.set_xlabel('Compostos', fontsize = 10)
ax.set_ylabel('Quantidade', fontsize = 10)
ax.set_title('Top 5 Compostos mais utilizados')
plt.show()
```

# **02: Plotar os histogramas com seaborn**

```python
dados.describe()
```

```python
dados['g-0'].hist()
plt.show()
```

```python
plt.figure(figsize = (15, 10))
dados['g-0'].hist(bins = 50)
plt.show()
```

```python
fig, ax = plt.subplots(2, constrained_layout = True, figsize = (15, 8))

ax1 = sns.histplot(data = dados.loc[:, 'g-0': 'g-771'].describe().T['mean'], bins = 50, ax=ax[0], color = 'blue')
ax[0].set_xlabel('M√©dia dos G', fontsize = 10)
ax[0].set_ylabel('Quantidade', fontsize = 10)
ax[0].set_title('Histograma do G', fontsize = 20)

ax1 = sns.histplot(data = dados.loc[:, 'c-0': 'c-99'].describe().T['mean'], bins = 50, ax=ax[1], color = 'purple')
ax[1].set_xlabel('M√©dia dos C', fontsize = 10)
ax[1].set_ylabel('Quantidade dos C', fontsize = 10)
ax[1].set_title('Histograma dos C', fontsize = 20)

plt.show()
```

O comando fig, ax = plt.subplots(2, constrained_layout=True, figsize =(20,8)) nos faz primeiro dizer que √© uma fig, ap√≥s o nome do array, com isso voc√™ cria subplots e passa por parametro a quantidade de plots, o constrained_layout=True para ajuste e p figsize para o tamanho do histograma.

# **05: Fazer outras an√°lises com o boxplot e histogramas**

```python
fig, ax = plt.subplots(3, constrained_layout = True, figsize = (10, 15))

ax1 = sns.boxplot(x = 'g-5', y = 'tratamento', data = dados, ax=ax[0], color = 'slateblue')
ax[0].set_xlabel('g-5', fontsize = 15)
ax[0].set_ylabel('Tratamentos', fontsize = 15)
ax[0].set_title('Tratamentos', fontsize = 20)

ax2 = sns.boxplot(x = 'g-5', y = 'dose', data = dados, ax=ax[1], color = 'gold')
ax[1].set_xlabel('g-5', fontsize = 15)
ax[1].set_ylabel('Doses', fontsize = 15)
ax[1].set_title('Doses', fontsize = 20)

ax3 = sns.boxplot(x = 'tempo', y = 'g-5', data = dados, ax=ax[2], color = 'turquoise')
ax[2].set_xlabel('Tempo', fontsize = 15)
ax[2].set_ylabel('g-5', fontsize = 15)
ax[2].set_title('Tempo', fontsize = 20)
plt.show()
```

Acima podemos ver os boxplots do tratamento, dose e tempo, respectivamente.
